3.3.1. Числа

Интерпретатор Python можно использовать как простой калькулятор: мы вводим 
в него выражение, а он возвращает значение. Синтаксис выражений прост: арифметические операторы +, -, * и / работают так же, как в любой компьютерной программе, которую вы могли использовать, например, в цифровом калькуляторе или 
электронной таблице Excel; круглые скобки (), как обычно, могут использоваться 
для группировки. Например: 


```python
2 + 2
```

```python
(50 - 5 * 6) / 4
```

```python
# результат деления всегда с плавающей точкой
8 / 5
```

Целые числа (например, 2, 4, 20) имеют тип int, а числа с дробной частью (например, 5.0, 1 .6) имеют тип float. О числовых типах мы поговорим позднее.

Оператор деления / всегда возвращает float. Чтобы выполнить деление нацело и 
получить целочисленный результат (отбрасив дробную часть), вы можете использовать оператор //, а для вычисления остатка - оператор %.

```python
# обычное деление с плавающей точкой
17 / 3
```

```python
# целочисленное деление
17 // 3
```

```python
# остаток от деления
17 % 3
```

```python
# проверка: результат * делитель + остаток
5 * 3 + 2
```

Для вычисления степени можно использовать оператор **. 

```python
# 5 в квадрате
5**2
```

```python
# 2 в седьмой степени
2**7
```
